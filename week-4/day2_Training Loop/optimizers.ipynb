```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
model = nn.Sequential(
    nn.Linear(10, 50),
    nn.ReLU(),
    nn.Linear(50, 2)
)
```

```python
sgd_optimizer = optim.SGD(model.parameters(), lr=0.01)

sgd_momentum_optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

sgd_momentum_optimizer.zero_grad()
dummy_loss_sgd = model(torch.randn(1, 10)).sum()
dummy_loss_sgd.backward()
sgd_momentum_optimizer.step()
```

```python
adam_optimizer = optim.Adam(model.parameters(), lr=0.001)

adamw_optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=0.01)

adam_optimizer.zero_grad()
dummy_loss_adam = model(torch.randn(1, 10)).sum()
dummy_loss_adam.backward()
adam_optimizer.step()
```
