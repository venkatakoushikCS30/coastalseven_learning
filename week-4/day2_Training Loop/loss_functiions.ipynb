```python
import torch
import torch.nn as nn
```

```python
mse_loss_fn = nn.MSELoss()
predictions_mse = torch.tensor([2.5, 5.1, 7.8], requires_grad=True)
targets_mse = torch.tensor([3.0, 5.0, 8.0])

mse_loss = mse_loss_fn(predictions_mse, targets_mse)
print(f"MSE Loss: {mse_loss.item():.4f}")
```

```python
bce_loss_fn = nn.BCELoss()

predictions_bce = torch.tensor([0.9, 0.1, 0.8], requires_grad=True)
targets_bce = torch.tensor([1.0, 0.0, 1.0])

bce_loss = bce_loss_fn(predictions_bce, targets_bce)
print(f"BCE Loss: {bce_loss.item():.4f}")
```

```python
ce_loss_fn = nn.CrossEntropyLoss()
```

```python
logits_ce = torch.tensor([
    [2.5, 0.3, 0.1],
    [0.1, 2.8, 0.5]
], requires_grad=True)

targets_ce = torch.tensor([0, 1])

ce_loss = ce_loss_fn(logits_ce, targets_ce)
print(f"Cross Entropy Loss: {ce_loss.item():.4f}")
```
