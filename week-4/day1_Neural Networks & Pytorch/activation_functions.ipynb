```python
import numpy as np
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stability trick
    return exp_x / np.sum(exp_x)
```

```python

x = np.array([-2, -1, 0, 1, 2])

print("Input:")
print(x)
```

***1. ReLu***

```python
print("\nReLU Output:")
print(relu(x))
```

***2. Sigmoid***

```python
print("\nSigmoid Output:")
print(sigmoid(x))
```

***3. Tanh***

```python
print("\nTanh Output:")
print(tanh(x))
```

***4. SoftMax***

```python
print("\nSoftmax Output:")
print(softmax(x))

print("\nSum of Softmax Probabilities:")
print(np.sum(softmax(x)))
```
