```python
import torch
import torch.nn as nn
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()

        # Convolution Layer
        # in_channels=1 (grayscale image)
        # out_channels=1 (one feature map)
        # kernel_size=3
        # stride=1
        # padding=1
        self.conv = nn.Conv2d(
            in_channels=1,
            out_channels=1,
            kernel_size=3,
            stride=1,
            padding=1
        )

        # Max Pooling Layer
        self.pool = nn.MaxPool2d(
            kernel_size=2,
            stride=2
        )

    def forward(self, x):
        print("Input Shape:", x.shape)

        x = self.conv(x)
        print("After Convolution:", x.shape)

        x = self.pool(x)
        print("After Pooling:", x.shape)

        return x
```

```python
model = SimpleCNN()
```

```python
x = torch.randn(1, 1, 28, 28)
```
