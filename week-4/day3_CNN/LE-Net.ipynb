```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
(X_train ,y_train) , ( X_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

print(X_train.shape)
```

```python
model = models.Sequential([
    layers.Conv2D(
        filters = 6,
        kernel_size = (5,5),
        activation = 'tanh',
        input_shape = (28,28,1)
    ),
    layers.AveragePooling2D(pool_size=(2,2)),

    layers.Conv2D(
        filters = 16,
        kernel_size = (5,5),
        activation = 'tanh'
    ) ,
    layers.AveragePooling2D(pool_size=(2,2)),

    layers.Flatten(), #Flatten() converts the matrix into 1D array ; so that the fully connected layer can process the data.

    #fully connected layer:

    layers.Dense(
        120,
        activation = 'tanh',
    ),
    layers.Dense(
        84,
        activation = 'tanh', #this and the above are hidden layers
    ),
    layers.Dense(
        10,
        activation = 'softmax', #this is o/p layer
    )
]

)
```

```python
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']

)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs = 10,
    batch_size = 64,
    validation_split = 0.1
)
```

```python
test_loss , test_acc = model.evaluate(
    X_test,
    y_test
)

test_acc , test_loss
```
