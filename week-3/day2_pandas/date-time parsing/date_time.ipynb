```python
import pandas as pd

dates = pd.Series(["2025-01-10", "2025-02-15", "2025-03-20"])
type(dates)
```

**1. Basic Parsing with pd.to_datetime()**

```python
parsed_dates = pd.to_datetime(dates)

print(parsed_dates)
print(parsed_dates.dtype)

type(parsed_dates)
```

**2. Different Date Formats**

```python
dates = pd.Series([
    "10-01-2025",
    "15-02-2025",
    "20-03-2025"
])

pd.to_datetime(dates, format="%d-%m-%Y")
```

| Code | Meaning        |
| ---- | -------------- |
| `%Y` | 4-digit year   |
| `%y` | 2-digit year   |
| `%m` | Month          |
| `%d` | Day            |
| `%H` | Hour (24-hour) |
| `%M` | Minute         |
| `%S` | Second         |


**4. Parsing a DataFrame Column**

```python
df = pd.DataFrame({
    "Date": ["2025-01-01", "2025-02-10", "2025-03-15"]
})

df["Date"] = pd.to_datetime(df["Date"])

print(df.dtypes)
```

**5. Working with Date and Time**

```python
date = pd.to_datetime("2026-06-02 14:30:45")
```

```python
date.hour
```

```python
date.minute
```

```python
date.second
```

**6. Date Filtering**

```python
df = pd.DataFrame({
    "Date": pd.to_datetime([
        "2025-01-01",
        "2025-02-01",
        "2025-03-01"
    ])
})

df[df["Date"] > "2025-01-15"]
```

**7. Date Arithmetic**

```python
df["Date"] + pd.Timedelta(days=7)
```

```python
df["Date"] - pd.Timedelta(days=30)
```

**8.Calculate Difference Between Dates**

```python
start = pd.to_datetime("2025-01-01")
end = pd.to_datetime("2025-01-15")

difference = end - start

print(difference)
```
