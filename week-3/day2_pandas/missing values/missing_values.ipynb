```python
df.isna() #checking for missing values using isna() method
```

```python
df.isnull().sum() #counting the number of missing values in each column using isnull() method and sum() function
```

```python
df.fillna(0) #filling missing values with 0 using fillna() method
```

```python
df["Age"] = df["Age"].fillna(30) #filling specific column
```

```python
df
```

```python
df.fillna({
    "Name": "Unknown",
    "Age": 30,
    "Salary": 0
}) #Different Values for Different Columns
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean()) #filling missing values with mean of the column
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median()) #filling missing values with median of the column
```

```python
df["Name"] = df["Name"].fillna(df["Name"].mode()[0]) #filling missing values with mode of the column
```

```python
df.ffill() #forward fill method to fill missing values with the previous value in the column
```

```python
df.bfill() #backward fill method to fill missing values with the next value in the column
```
