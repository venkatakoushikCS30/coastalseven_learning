*join()*

```python
import pandas as pd

df1 = pd.DataFrame({
    "name": ["John", "Alice"]
}, index=[1, 2])

df2 = pd.DataFrame({
    "salary": [50000, 60000]
}, index=[1, 2])

print(df1)
print(df2)
```

```python
result = df1.join(df2)

print(result)
```

*1. Left Join with Indexes*

```python
df1 = pd.DataFrame({
    "name": ["John", "Alice", "Bob"]
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    "salary": [50000, 60000]
}, index=[1, 2])

result = df1.join(df2)

print(result)
```

*2. Inner Join*

```python
result = df1.join(
    df2,
    how="inner"
)

print(result)
```

*3. Join Multiple DataFrames*

```python
students = pd.DataFrame({
    "name": ["John", "Alice"]
}, index=[1, 2])

marks = pd.DataFrame({
    "math": [90, 85]
}, index=[1, 2])

attendance = pd.DataFrame({
    "attendance": [95, 92]
}, index=[1, 2])

result = students.join([marks, attendance])

print(result)
```
