```python
import pandas as pd
```

*1. Basic Merge*

```python
employees = pd.DataFrame({
    "emp_id": [1, 2, 3],
    "name": ["John", "Alice", "Bob"]
})

print(employees)
```

```python
salaries = pd.DataFrame({
    "emp_id": [1, 2, 3],
    "salary": [50000, 60000, 70000]
})

print(salaries)
```

```python
result = pd.merge(employees, salaries, on="emp_id")
print(result)
```

*2. Left Join*

```python
employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4],
    "name": ["John", "Alice", "Bob", "David"]
})

salaries = pd.DataFrame({
    "emp_id": [1, 2],
    "salary": [50000, 60000]
})

result = pd.merge(
    employees,
    salaries,
    on="emp_id",
    how="left"
)

print(result)
```

*3. Merge Using Different Column Names*

```python
users = pd.DataFrame({
    "user_id": [1, 2, 3],
    "name": ["John", "Alice", "Bob"]
})

orders = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "amount": [100, 200, 300]
})

result = pd.merge(
    users,
    orders,
    left_on="user_id",
    right_on="customer_id"
)

print(result)
```

*4. Merge on Multiple Columns*

```python
sales = pd.DataFrame({
    "product": ["Laptop", "Laptop", "Phone"],
    "region": ["East", "West", "East"],
    "sales": [100, 150, 80]
})

targets = pd.DataFrame({
    "product": ["Laptop", "Laptop", "Phone"],
    "region": ["East", "West", "East"],
    "target": [120, 160, 90]
})

result = pd.merge(
    sales,
    targets,
    on=["product", "region"]
)

print(result)
```
