```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
```

```python
iris = load_iris()
```

```python
X = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)
```

```python
print("=" * 50)
print("ORIGINAL DATA")
print("=" * 50)
print(X.head())
```

```python
# ---------------------------
# StandardScaler
# ---------------------------
standard_scaler = StandardScaler()
X_standard = standard_scaler.fit_transform(X)

X_standard = pd.DataFrame(
    X_standard,
    columns=iris.feature_names
)
```

```python
print("\n" + "=" * 50)
print("STANDARD SCALER OUTPUT")
print("=" * 50)
print(X_standard.head())
```

```python
# Check mean and std
print("\nColumn Means:")
print(X_standard.mean())
```

```python
print("\nColumn Standard Deviations:")
print(X_standard.std())
```

```python
# ---------------------------
# MinMaxScaler
# ---------------------------
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X)

X_minmax = pd.DataFrame(
    X_minmax,
    columns=iris.feature_names
)
```

```python

print("\n" + "=" * 50)
print("MIN-MAX SCALER OUTPUT")
print("=" * 50)
print(X_minmax.head())
```

```python
# Check min and max values
print("\nColumn Minimums:")
print(X_minmax.min())
```

```python
print("\nColumn Maximums:")
print(X_minmax.max())
```

```python
# ---------------------------
# Compare First 5 Rows
# ---------------------------
comparison = pd.concat(
    [
        X.head(),
        X_standard.head().add_prefix("std_"),
        X_minmax.head().add_prefix("minmax_")
    ],
    axis=1
)
```

```python
print("\n" + "=" * 50)
print("COMPARISON OF ORIGINAL VS SCALED DATA")
print("=" * 50)
print(comparison)
```
