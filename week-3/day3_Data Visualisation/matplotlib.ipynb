```python
import matplotlib.pyplot as plt
```

**MatplotLib Basics**

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y)
plt.show()
```

```python
x=["may","june","july","aug","sep"]
y=[10,20,30,40,50]
plt.xlabel("Months")
plt.ylabel("Sales")
plt.plot(x,y,label="sales")
plt.legend()
plt.grid(True)
plt.figure(figsize=(10,5))
plt.show()
```

***1. Line Plot*** 

Used to show trends over time.

```python
months = ["may","june","july","aug"]
sales1 = [100,120,135,170]
plt.plot(months,sales1 , color="red" , linestyle="--" , marker = "o" , linewidth=2) #customize the plot
plt.grid(True)
plt.show()
```

##

*Multiple Lines*

```python
sales2 = [150,180,200,220]
plt.plot(months,sales1 ,color="red" , marker ="o",label = "Store A")
plt.plot(months,sales2 , color="blue" ,marker = "o", label = "Store B")
plt.legend()
plt.show()
```

***2.Bar Plot***

Compare categories.

```python
courses = ["Python","Java","C++"]
students = [120 , 90 , 70]
plt.bar(courses ,students)
plt.show()
```

2 types of bar plot - horizontal bar plot and vertical bar plot

```python
plt.barh(courses, students) #barh - horizontal bar
```

```python
plt.bar(courses,students , color = "green") #bar - vertical bar
plt.title("Courses vs Students")
plt.xlabel("Courses")
plt.ylabel("Students")  
plt.show()
```

***3. Scatter Plots***

Find relationships between variables.

```python
age = [20,21,22,23,25]
salaries = [15000,24000,25000,32000,38000]
plt.scatter(age,salaries)
plt.show()
```

here if age increase and salary increase - so it is a positive relationship

2 plots on the same figure

```python
import numpy as np
x = np.array([23,24,25,26])
y= np.array([15000,24000,20000,32000])
plt.scatter(x,y)

x = np.array([27,28,29])
y= np.array([15000,10000,32000])
plt.scatter(x,y)
plt.show()
```

***4. Histogram***

Understand distribution of data.

```python
marks = [45,50,52,60,65,70,72,75,80,85,90]
plt.hist(marks,bins = 5)
plt.show()
```

Histogram is used to check:
Normal distribution,

Frequency

***5 .Box Plot***

purpose - to detect median , quatriles and outliers

```python
marks = [20,30,40,50,60,70,80,90,200] #here 200 is the outlier

plt.boxplot(marks)

plt.show()
```

plot customization

```python
linestyle = dict(
     marker =".",
         markersize = 20,
         markerfacecolor = "#1cd3fc",
         markeredgecolor = "#1cd3fc",
         linestyle = "solid",
         linewidth = 4,
)
```

```python
x = np.array([2023,2024,2025,2026])
y1 = np.array([15,25,30,20])
y2 = np.array([17,23,38,5])
y3 = np.array([13,15,20,30])
plt.plot(x,y1, color = "#1c5bfc",**linestyle)
plt.plot(x,y2, color = "#1cfc45",**linestyle)
plt.plot(x,y3, color = "#fc1c1c",**linestyle)
plt.title("Sales vs Years" , fontsize = 20)
plt.xlabel("Years" , fontsize = 15)
plt.ylabel("Sales", fontsize = 15)
plt.xticks(x)
```

***6. Pie Charts***

```python
categories = ["Freshmen","Sophomore","Junior","Senior"]
values = np.array([300,250,275,225])
colors = ["red" , "yellow","blue","green"]

plt.pie(values, labels = categories,
        autopct = "%1.1f%%",
        colors = colors,
        explode =[0,0,0,0.1],
        shadow = True,
         )
plt.title("Student Demographics", fontsize = 20)
plt.show()
```

***7. Sub Plots***

```python
figure , axes = plt.subplots(2,2) #subplot - 2 rows and 2 columns
x = np.array([1,2,3,4,5])
axes[0,0].plot(x,x*2 , color = "red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2,color = "blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x,x**3,color = "green")
axes[1,0].set_title("x**3")

axes[1,1].plot(x,x**4,color = "purple")
axes[1,1].set_title("x**4")

plt.tight_layout()
plt.show()
```
