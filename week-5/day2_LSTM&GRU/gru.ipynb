```python
import torch
import torch.nn as nn
import numpy as np
```

```python
X = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [3, 4, 5],
    [4, 5, 6]
], dtype=np.float32)

y = np.array([
    [4],
    [5],
    [6],
    [7]
], dtype=np.float32)

# Convert to tensors
X = torch.tensor(X)
y = torch.tensor(y)

# Reshape for GRU
# (batch_size, sequence_length, input_size)
X = X.unsqueeze(-1)

print("Input Shape:", X.shape)
```

```python
class GRUNetwork(nn.Module):
    def __init__(self):
        super().__init__()

        self.gru = nn.GRU(
            input_size=1,
            hidden_size=16,
            batch_first=True
        )

        self.fc = nn.Linear(16, 1)

    def forward(self, x):

        output, hidden = self.gru(x)

        # Last time step output
        out = output[:, -1, :]

        out = self.fc(out)

        return out

model = GRUNetwork()

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 500

for epoch in range(epochs):

    pred = model(X)

    loss = criterion(pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.4f}")
```

```python
test_seq = torch.tensor(
    [[[5.0], [6.0], [7.0]]]
)

prediction = model(test_seq)

print("\nPredicted next value:")
print(prediction.item())
```
