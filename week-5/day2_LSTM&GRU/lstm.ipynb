```python
import torch
import torch.nn as nn
```

```python
x = torch.tensor([[[1.0],
                   [2.0],
                   [3.0]]])
```

```python

lstm = nn.LSTM(
    input_size=1,
    hidden_size=4,
    num_layers=1,
    batch_first=True
)
```

```python
output, (hidden_state, cell_state) = lstm(x)
```

```python
print("Output:")
print(output)
```

```python
print("\nFinal Hidden State (hn):")
print(hn)
```
